```python
import aiohttp
import asyncio
from bs4 import BeautifulSoup
import re

async def fetch_html(url, session):
    async with session.get(url) as response:
        return await response.text()

async def async_html_loader(urls):
    async with aiohttp.ClientSession() as session:
        tasks = [fetch_html(url, session) for url in urls]
        return await asyncio.gather(*tasks)

def remove_header_footer(html):
    soup = BeautifulSoup(html, 'html.parser')
    header = soup.find('header')
    footer = soup.find('footer')
    if header:
        header.decompose()
    if footer:
        footer.decompose()
    return soup

def html2text(html):
    # Remove header and footer
    soup = remove_header_footer(html)
    text = soup.get_text()
    # Replace three or more consecutive newlines with a single newline
    text = re.sub(r'\n{3,}', '\n', text)
    return text

async def main(urls):
    # Step 1: Load HTML content asynchronously
    html_contents = await async_html_loader(urls)
    
    # Step 2: Convert HTML to text and strip extra newlines
    text_contents = [html2text(html) for html in html_contents]
    
    # Write text contents to a file
    with open('extracted_text.txt', 'w', encoding='utf-8') as file:
        for text in text_contents:
            file.write(text)
            file.write('\n' + '-' * 80 + '\n')

    print("Text content has been saved to 'extracted_text.txt'.")

# List of URLs to fetch
urls = ["https://www.honda.co.uk/cars/owners/my-honda-plus/privacy-policy.html"]


# Run the full workflow
await main(urls)
```
